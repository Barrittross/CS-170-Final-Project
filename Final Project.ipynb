```python
# This code defines a helper function to extract numeric values from car prices,
# cleans and converts them into millions of USD, selects the top 10 most expensive cars,
# and visualizes them in a bar graph with rotated labels for readability.
import re 
data = pd.read_csv(
    r"C:\Users\barri\OneDrive - Ursinus College\Desktop\Python-Projects\Cars Datasets 2025.csv",
    encoding= 'latin-1'
)
# Function to extract numeric price values
def extract_price(s):
    nums = re.findall(r"\d[\d,]*", str(s))
    if not nums:
        return None
    return float(nums[-1].replace(",", ""))  # take highest number

# Clean the price column
data["PriceNum"] = data["Cars Prices"].apply(extract_price)

# Convert to millions
data["PriceMillions"] = data["PriceNum"] / 1_000_000

# Get top 10 most expensive cars
top10 = data.sort_values("PriceNum", ascending=False).head(10)

# Plot
plt.figure(figsize=(10,5))
plt.bar(top10["Cars Names"].astype(str), top10["PriceMillions"])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Car")
plt.ylabel("Price (Millions of USD)")
plt.title("Top 10 Most Expensive Cars")
plt.tight_layout()
plt.show()
```

# Fastest Cars 

```python
# This shows the first 2 rows of the data set
data.head(2)
```

```python
# This code shows the performance of the top 5 fastest cars
# Ensure the performance column is numeric
data['Performance(0 - 100 )KM/H'] = pd.to_numeric(data['Performance(0 - 100 )KM/H'], errors='coerce')

# Group by car name, take the minimum acceleration time (fastest), and sort ascending
fastest_cars = (
    data.groupby('Cars Names')['Performance(0 - 100 )KM/H']
    .min()
    .sort_values(ascending=True)
    .head(5)
)

print(fastest_cars)
```

```python
# This code converts the acceleration column to numeric, finds the 5 fastest cars by 0–100 km/h time,
# and plots them in a bar chart with distinct colors and rotated labels for clarity.
data['Performance(0 - 100 )KM/H'] = pd.to_numeric(data['Performance(0 - 100 )KM/H'], errors='coerce')

# Get the 5 fastest cars
fastest_5 = (
    data.groupby('Cars Names')['Performance(0 - 100 )KM/H']
    .min()
    .sort_values()
    .head(5)
    .reset_index()
)

# Plotting
plt.figure(figsize=(9, 4))
plt.bar(fastest_5['Cars Names'], fastest_5['Performance(0 - 100 )KM/H'],
        color=['red', 'blue', 'green', 'orange', 'cyan'], width=0.5)
plt.xlabel('Cars Names')
plt.ylabel('Performance (0 - 100) KM/H (seconds)')
plt.title('Top 5 Fastest Cars by Acceleration')

# Rotate x-axis labels to avoid overlap
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()

```

```python
# This code imports libraries for data analysis and visualization, sets plotting styles, 
# and loads the Cars Datasets 2025 CSV file into a pandas DataFrame for exploration.
import pandas as pd # data analysis and manipulation (DF)
import numpy as np # numerical operations
import seaborn as sns # creates statistical graphics 

import matplotlib.pyplot as plt 
import matplotlib.mlab as mlab
import matplotlib # data visualization 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create

# reading data
data = pd.read_csv(
    r"C:\Users\barri\OneDrive - Ursinus College\Desktop\Python-Projects\Cars Datasets 2025.csv",
    encoding= 'latin-1'
)
```

```python
# Scatter Plot showing Horsepower vs Car Price 
# clean price column (remove $ and commas, convert to float)
data["Cars Prices"] = data["Cars Prices"].astype(str) \
    .replace(r'[\$,]', '', regex=True) \
    .replace(',', '', regex=True)

data["Cars Prices"] = data["Cars Prices"].str.extract(r'(\d+\.?\d*)')[0].astype(float)

# clean horsepower column (remove non-numeric entries if any)
data["HorsePower"] = data["HorsePower"].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)

# scatter plot
plt.scatter(data["HorsePower"], data["Cars Prices"])
plt.title("Horsepower vs Car Price")
plt.xlabel("Horsepower")
plt.ylabel("Car Price (USD)")
plt.show()
```
